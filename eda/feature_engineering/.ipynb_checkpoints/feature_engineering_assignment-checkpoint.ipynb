```python
import numpy as np

# Calculate IQR
Q1 = raisins_numeric_df.quantile(0.25)
Q3 = raisins_numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using IQR
outliers_iqr = (raisins_numeric_df < (Q1 - 1.5 * IQR)) | (raisins_numeric_df > (Q3 + 1.5 * IQR))

# Calculate Z-score
z_scores = np.abs((raisins_numeric_df - raisins_numeric_df.mean()) / raisins_numeric_df.std())

# Identify outliers using Z-score
outliers_zscore = z_scores > 3  # Threshold Z-score of 3

# Combine results
outliers = outliers_iqr | outliers_zscore

# Display outliers
print("Outliers using both IQR and Z-score methods:")
outliers_df = raisins_numeric_df[outliers]
print(outliers_df)
```

### Explore sklearn’s module for outlier detection (eg. LOF, OneClassSVM)

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
import numpy as np

# Generate sample data
np.random.seed(42)
X = 0.3 * np.random.randn(100, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
X = np.vstack([X, X_outliers])

# Fit the model: Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_lof = lof.fit_predict(X)

# Fit the model: One-Class SVM
svm = OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)
svm.fit(X)
y_svm = svm.predict(X)

# Display results
print("Local Outlier Factor (LOF) predictions:")
print(y_lof)

print("\nOne-Class SVM predictions:")
print(y_svm)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

# Define the target column
target_column = 'Class'  # Replace 'target_column' with your actual target column name

# Separate features and target
X = raisins_numeric_df.drop(columns=[target_column])
y = raisins_numeric_df[target_column]

# Use SelectKBest with f_regression to select top k features
selector = SelectKBest(score_func=f_regression, k=5)  # Adjust k as needed
X_selected = selector.fit_transform(X, y)

# Get selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get names of selected features
selected_feature_names = X.columns[selected_feature_indices]

# Display selected feature names
print("Selected features using Filter Method:")
print(selected_feature_names)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Define the estimator (model) for feature selection
estimator = LinearRegression()  # Replace with your preferred estimator

# Perform Recursive Feature Elimination (RFE) without specifying n_features_to_select in the constructor
selector = RFE(estimator)
selector = selector.fit(X, y)

# Set the number of features to select using the n_features_to_select attribute
selector.n_features_to_select = 5  # Adjust as needed

# Get selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get names of selected features
selected_feature_names = X.columns[selected_feature_indices]

# Display selected feature names
print("Selected features using Wrapper Method:")
print(selected_feature_names)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Assuming X is your feature matrix

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Perform t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

# Visualize PCA results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.5)
plt.title('PCA')

# Visualize t-SNE results
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', alpha=0.5)
plt.title('t-SNE')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

# Create a DataFrame with the original features
data = {
    'feature1': [1, 2, 3, 4, 5],
    'feature2': [6, 7, 8, 9, 10]
}
df = pd.DataFrame(data)

# Perform feature extraction using PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df)

# Generate feature names manually
poly_columns = poly.get_feature_names_out(df.columns)

# Create a DataFrame with the new polynomial features
df_poly = pd.DataFrame(poly_features, columns=poly_columns)

# Perform feature extraction by creating interaction terms
interaction_features = df['feature1'] * df['feature2']
df['interaction'] = interaction_features

# Display the DataFrame with the new features
print("DataFrame with new polynomial features:")
print(df_poly)

print("\nDataFrame with interaction term:")
print(df)
```

```python
from imblearn.over_sampling import RandomOverSampler

X = raisins_numeric_df.drop(columns=['Class'])
y = raisins_numeric_df['Class']

# Perform Random Oversampling
oversampler = RandomOverSampler(random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Check the distribution of resampled target labels
resampled_target_distribution = pd.Series(y_resampled).value_counts(normalize=True)
print("\nResampled Target Label Distribution:")
print(resampled_target_distribution)
```
