## Group 1

### Perform missing data handling for columns with null values

```python
# Check for null values in each column
null_columns = athlete_df.columns[athlete_df.isnull().any()]

# Display columns with null values
print("Columns with null values:")
print(null_columns)
```

#### Replacing null values for 'Age' column

```python
age_np = np.array(athlete_df['Age'].dropna())
kde = KernelDensity(bandwidth=10, kernel='gaussian')
kde.fit(age_np[:, np.newaxis])  # sklearn expects data in 2D array format

# Generate points for plotting the KDE
x_plot = np.linspace(min(age_np), max(age_np), 1000)
log_dens = kde.score_samples(x_plot[:, np.newaxis])

# Plot the data and the KDE
plt.figure(figsize=(8, 6))
sns.histplot(athlete_df['Age'], bins=30, kde=False, alpha=0.5, label='Histogram of data', stat = 'density')
plt.plot(x_plot, np.exp(log_dens), label='Kernel Density Estimation', color = 'Red')
plt.title('Kernel Density Estimation (KDE) Age of Athletes')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Plot box plot using Seaborn
sns.boxplot(data=athlete_df, y='Age')

# Add labels
plt.title('Box Plot of Age')
plt.ylabel('Values')

# Show the plot
plt.show()
```

##### The 'Age' column seem to be postively skewed so we shall replace the NaN values with the median

```python
athlete_df['Age'].fillna(athlete_df['Age'].median(), inplace = True)
```

#### Replacing null values in 'Height' column

```python
height_np = np.array(athlete_df['Height'].dropna())
kde = KernelDensity(bandwidth=10, kernel='gaussian')
kde.fit(height_np[:, np.newaxis])  # sklearn expects data in 2D array format

# Generate points for plotting the KDE
x_plot = np.linspace(min(height_np), max(height_np), 1000)
log_dens = kde.score_samples(x_plot[:, np.newaxis])

# Plot the data and the KDE
plt.figure(figsize=(8, 6))
sns.histplot(athlete_df['Height'], bins=30, kde=False, alpha=0.5, label='Histogram of data', stat = 'density')
plt.plot(x_plot, np.exp(log_dens), label='Kernel Density Estimation', color = "Red")
plt.title('Kernel Density Estimation (KDE) Height of Athletes')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Plot box plot using Seaborn
sns.boxplot(data=athlete_df, y='Height')

# Add labels
plt.title('Box Plot of Height')
plt.ylabel('Values')

# Show the plot
plt.show()
```

##### The 'Height' column of the dataset is normally distributed so we shall replace the null values with the mean of the 'Height' column

```python
athlete_df['Height'].fillna(athlete_df['Height'].mean(), inplace = True)
```

```python
# Plot box plot using Seaborn
sns.boxplot(data=athlete_df, y='Weight')

# Add labels
plt.title('Box Plot of Weight')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# use .copy() to copy df so that the original dataframe is not effected
is_host_df = athlete_df.copy()

"""
'Is_Host' will be True when the team is hosting that year, for eg Brazil hosted in 2016 so 'Is_Host' will be labelled True for
Brazilian teams that participated in 2016, but labelled False for Brazilian teams that participated in other years
"""
is_host_df['Is_Host'] = False

# Mark the rows where the country is a host for the respective year
for year, host_country in host_countries.items():
    # Filter data for the current year
    is_host_df.loc[(is_host_df['Team'] == host_country) & (is_host_df['Year'] == int(year)), 'Is_Host'] = True
    

# Filter the df to include only host country rows
when_hosting_df = is_host_df[is_host_df['Is_Host'] == True]
# Filter the DataFrame to include only rows where the country has hosted once and they years in which they are not hosting
when_not_hosting_df = is_host_df[(is_host_df['Is_Host'] == False) & (is_host_df['Team'].isin(host_countries.values()))]
    
# Group the DataFrame by country and year, and calculate the total number of medals won in each year for each country when they were not hosting
total_medals_per_country_year_not_hosting = when_not_hosting_df.groupby(['Team', 'Year'])['Medal'].value_counts().unstack(fill_value = 0)
# Drop the 'NA' column if it exists
total_medals_per_country_year_not_hosting.drop(columns = 'NA', errors = 'ignore', inplace = True)
total_medals_per_country_year_not_hosting['Total'] = total_medals_per_country_year_not_hosting['Gold'] + total_medals_per_country_year_not_hosting['Bronze'] + total_medals_per_country_year_not_hosting['Silver']
total_medals_per_country_year_not_hosting.reset_index(inplace = True)

# Calculate the average number of medals won per year for each country when not hosting
average_medals_per_year_not_hosting_by_country = total_medals_per_country_year_not_hosting.groupby('Team')['Total'].mean().reset_index()
average_medals_per_year_not_hosting_by_country.columns = ['Country', 'Average_Medals_Per_Year_Not_Hosting']


# Group the DataFrame by country and year, and calculate the total number of medals won in each year for each country when they were hosting
total_medals_per_country_year_hosting = when_hosting_df.groupby(['Team', 'Year'])['Medal'].value_counts().unstack(fill_value=0)

# Drop the 'NA' column if it exists
total_medals_per_country_year_hosting.drop(columns='NA', errors='ignore', inplace=True)

# Calculate the total number of medals (Gold, Silver, Bronze) won per year for each country
total_medals_per_country_year_hosting['Total'] = total_medals_per_country_year_hosting['Gold'] + total_medals_per_country_year_hosting['Silver'] + total_medals_per_country_year_hosting['Bronze']

# Reset the index to make 'Team' a column again
total_medals_per_country_year_hosting.reset_index(inplace=True)

# Calculate the average number of medals won per year for each country
average_medals_per_year_hosting_by_country = total_medals_per_country_year_hosting.groupby('Team')['Total'].mean().reset_index()
average_medals_per_year_hosting_by_country.columns = ['Country', 'Average_Medals_Per_Year_Hosting']

```

```python
# Extract the total number of medals won per year when hosting and not hosting
medals_hosting = total_medals_per_country_year_hosting.groupby('Year')['Total'].sum()
medals_not_hosting = total_medals_per_country_year_not_hosting.groupby('Year')['Total'].sum()

# Perform the t-test
t_statistic, p_value = ttest_ind(medals_hosting, medals_not_hosting)

# Print the results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Interpret the results based on the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. Hosting the Olympics has a significant impact on the total number of medals won.")
else:
    print("Fail to reject the null hypothesis. There is no significant impact of hosting the Olympics on the total number of medals won.")
```
